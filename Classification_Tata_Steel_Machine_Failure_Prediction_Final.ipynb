## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Data Wrangling Code (No Data Leakage)

print("="*50)
print("🛠️ DATA WRANGLING OPERATIONS")
print("="*50)

# 1. Create a copy of the data
train_data = train_df.copy()
test_data = test_df.copy()

print("\n✅ Created working copies of datasets")

# 2. Drop unnecessary identifier columns
print("\n" + "="*50)
print("🔍 DROPPING IDENTIFIER COLUMNS")
print("="*50)

train_data.drop(['id', 'Product ID'], axis=1, inplace=True)
test_data.drop(['id', 'Product ID'], axis=1, inplace=True)

print(f"✅ Dropped 'id' and 'Product ID' columns")
print(f"Training data shape after dropping: {train_data.shape}")
print(f"Test data shape after dropping: {test_data.shape}")

# 3. Feature Engineering (creating new features)
print("\n" + "="*50)
print("✨ FEATURE ENGINEERING")
print("="*50)

# Temperature Difference (for Heat Dissipation Failure)
train_data['Temp_Difference'] = train_data['Process temperature [K]'] - train_data['Air temperature [K]']
test_data['Temp_Difference'] = test_data['Process temperature [K]'] - test_data['Air temperature [K]']
print("✅ Created 'Temp_Difference' feature")

# Power feature (Torque * Rotational Speed) / 1000 (for Power Failure)
train_data['Power'] = (train_data['Torque [Nm]'] * train_data['Rotational speed [rpm]']) / 1000
test_data['Power'] = (test_data['Torque [Nm]'] * test_data['Rotational speed [rpm]']) / 1000
print("✅ Created 'Power' feature")

# Stress product (Tool wear * Torque) for Overstrain Failure
train_data['Stress_Product'] = train_data['Tool wear [min]'] * train_data['Torque [Nm]']
test_data['Stress_Product'] = test_data['Tool wear [min]'] * test_data['Torque [Nm]']
print("✅ Created 'Stress_Product' feature")

# Rotational Speed squared
train_data['Rotational_speed_sq'] = train_data['Rotational speed [rpm]'] ** 2
test_data['Rotational_speed_sq'] = test_data['Rotational speed [rpm]'] ** 2
print("✅ Created 'Rotational_speed_sq' feature")

# 4. Display updated dataset info
print("\n" + "="*50)
print("📊 UPDATED DATASET INFORMATION")
print("="*50)
print(f"\nTraining data final shape: {train_data.shape}")
print(f"Test data final shape: {test_data.shape}")
print(f"\nAll columns in training data:")
print(list(train_data.columns))
print(f"\nAll columns in test data:")
print(list(test_data.columns))

# 5. Quick look at the new features
print("\n" + "="*50)
print("📈 NEW FEATURES - FIRST FEW ROWS")
print("="*50)
print("\nTraining Data - New Features Sample:")
display(train_data[['Temp_Difference', 'Power', 'Stress_Product', 'Rotational_speed_sq']].head())

print("\nTest Data - New Features Sample:")
display(test_data[['Temp_Difference', 'Power', 'Stress_Product', 'Rotational_speed_sq']].head())

# 6. Verify no data leakage
print("\n" + "="*50)
print("🔍 VERIFYING NO DATA LEAKAGE")
print("="*50)

failure_indicators = ['TWF', 'HDF', 'PWF', 'OSF', 'RNF', 'Machine failure']
print("Checking if any failure indicators are being used as features...")
print(f"Failure indicators in dataset: {failure_indicators}")
print(f"\nNote: These will be our TARGET variables, not features!")
print("✅ Data wrangling complete - ready for EDA and modeling")
````

## ***5. Hypothesis Testing***

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀): There is no significant difference in failure rates between Low-quality (Type L) and High-quality (Type H) products.

Alternate Hypothesis (H₁): Low-quality (Type L) products have a significantly higher failure rate than High-quality (Type H) products.

#### 2. Perform an appropriate statistical test.

````python
# ==================================================
# HYPOTHESIS TESTING - STATEMENT 1
# ==================================================

print("="*50)
print("📊 HYPOTHESIS TEST 1: Failure Rate by Product Type")
print("="*50)

# Extract data for Type L and Type H
type_l_failures = train_data[train_data['Type'] == 'L']['Machine failure']
type_h_failures = train_data[train_data['Type'] == 'H']['Machine failure']

# Calculate failure rates
type_l_rate = type_l_failures.mean() * 100
type_h_rate = type_h_failures.mean() * 100

print(f"\n📌 Sample Statistics:")
print(f"Type L (Low Quality):")
print(f"  Sample Size: {len(type_l_failures)}")
print(f"  Failures: {type_l_failures.sum()}")
print(f"  Failure Rate: {type_l_rate:.2f}%")
print(f"\nType H (High Quality):")
print(f"  Sample Size: {len(type_h_failures)}")
print(f"  Failures: {type_h_failures.sum()}")
print(f"  Failure Rate: {type_h_rate:.2f}%")

# Perform statistical test (Chi-square test for proportions)
from scipy.stats import chi2_contingency

# Create contingency table
contingency_table = pd.crosstab(train_data['Type'], train_data['Machine failure'])
contingency_table = contingency_table.loc[['L', 'H']]  # Only L and H

print(f"\n📊 Contingency Table (L vs H):")
print(contingency_table)

# Perform Chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"\n📊 Chi-square Test Results:")
print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")
print(f"Degrees of freedom: {dof}")

# Conclusion
alpha = 0.05
print(f"\n📌 Conclusion (at α = {alpha}):")
if p_value < alpha:
    print(f"❌ Reject Null Hypothesis (p-value = {p_value:.6f} < {alpha})")
    print("✅ There is statistically significant evidence that failure rates differ between Type L and Type H products.")
else:
    print(f"✅ Fail to Reject Null Hypothesis (p-value = {p_value:.6f} >= {alpha})")
    print("❌ There is NOT enough evidence to conclude failure rates differ between Type L and Type H products.")
````

##### Which statistical test have you done to obtain P-Value?

I performed a Chi-square test of independence using the contingency table comparing failure counts between Type L and Type H products.

##### Why did you choose the specific statistical test?

I chose the Chi-square test because:

Data Type: Both variables are categorical (Product Type and Machine Failure are binary/categorical)

Comparison: We're comparing proportions (failure rates) between two independent groups

Sample Size: Large sample sizes (95,354 and 8,923) make Chi-square test appropriate and reliable

Expected Frequencies: All expected frequencies are >5, satisfying the test assumption

Research Question: Testing association between two categorical variables is exactly what Chi-square test handles

Interpretation of Results:

P-value = 0.0091 which is less than 0.05

We reject the null hypothesis

Conclusion: There is statistically significant evidence that Low-quality products (Type L) have a higher failure rate (1.67%) than High-quality products (Type H) (1.30%)

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀): There is no significant difference in mean temperature difference between failure and non-failure cases.

Alternate Hypothesis (H₁): Failure cases have a significantly lower mean temperature difference than non-failure cases.

#### 2. Perform an appropriate statistical test.

````python
# ==================================================
# HYPOTHESIS TESTING - STATEMENT 3
# ==================================================

print("="*50)
print("📊 HYPOTHESIS TEST 3: Temperature Difference by Failure Status")
print("="*50)

# Extract data for failure and non-failure cases
temp_diff_failure = train_data[train_data['Machine failure'] == 1]['Temp_Difference']
temp_diff_no_failure = train_data[train_data['Machine failure'] == 0]['Temp_Difference']

# Calculate statistics
print(f"\n📌 Sample Statistics:")
print(f"\nNo Failure (0):")
print(f"  Sample Size: {len(temp_diff_no_failure)}")
print(f"  Mean Temp Difference: {temp_diff_no_failure.mean():.3f} K")
print(f"  Std Deviation: {temp_diff_no_failure.std():.3f} K")

print(f"\nFailure (1):")
print(f"  Sample Size: {len(temp_diff_failure)}")
print(f"  Mean Temp Difference: {temp_diff_failure.mean():.3f} K")
print(f"  Std Deviation: {temp_diff_failure.std():.3f} K")

# Check normality assumption with histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(temp_diff_no_failure, kde=True, ax=axes[0], color='green', bins=30)
axes[0].set_title('Temp Difference Distribution: No Failure', fontsize=12)
axes[0].set_xlabel('Temperature Difference [K]')

sns.histplot(temp_diff_failure, kde=True, ax=axes[1], color='red', bins=30)
axes[1].set_title('Temp Difference Distribution: Failure', fontsize=12)
axes[1].set_xlabel('Temperature Difference [K]')

plt.tight_layout()
plt.show()

# Perform statistical test (Mann-Whitney U test - non-parametric)
from scipy.stats import mannwhitneyu

statistic, p_value = mannwhitneyu(temp_diff_failure, temp_diff_no_failure, alternative='less')

print(f"\n📊 Mann-Whitney U Test Results:")
print(f"U-statistic: {statistic:.4f}")
print(f"P-value: {p_value:.10f}")

# Also perform independent t-test for comparison (despite non-normality)
from scipy.stats import ttest_ind

t_stat, t_p_value = ttest_ind(temp_diff_failure, temp_diff_no_failure, alternative='less')

print(f"\n📊 Independent t-test Results (for reference):")
print(f"t-statistic: {t_stat:.4f}")
print(f"P-value: {t_p_value:.10f}")

# Conclusion
alpha = 0.05
print(f"\n📌 Conclusion (at α = {alpha}) using Mann-Whitney U test:")
if p_value < alpha:
    print(f"❌ Reject Null Hypothesis (p-value = {p_value:.6f} < {alpha})")
    print("✅ There is statistically significant evidence that failure cases have LOWER temperature difference.")
else:
    print(f"✅ Fail to Reject Null Hypothesis (p-value = {p_value:.6f} >= {alpha})")
    print("❌ There is NOT enough evidence to conclude temperature difference is lower for failures.")

# Calculate mean difference
mean_diff = temp_diff_no_failure.mean() - temp_diff_failure.mean()
print(f"\n📈 Mean Temperature Difference Gap: {mean_diff:.3f} K")
print(f"Failures occur when temperature difference is {mean_diff:.3f}K lower on average")
````

Robust to Outliers: Non-parametric tests are less affected by outliers

### 2. Handling Outliers

````python
# ==================================================
# 6.2 HANDLING OUTLIERS
# ==================================================

print("="*50)
print("📊 6.2: OUTLIERS DETECTION")
print("="*50)

# Select numerical columns for outlier detection
numerical_cols = ['Air temperature [K]', 'Process temperature [K]',
                  'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]',
                  'Temp_Difference', 'Power', 'Stress_Product']

# Calculate IQR and detect outliers
outlier_info = {}

for col in numerical_cols:
    Q1 = train_data[col].quantile(0.25)
    Q3 = train_data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = train_data[(train_data[col] < lower_bound) | (train_data[col] > upper_bound)]
    outlier_pct = (len(outliers) / len(train_data)) * 100

    outlier_info[col] = {
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outlier Count': len(outliers),
        'Outlier Percentage': outlier_pct
    }

# Display outlier information
outlier_df = pd.DataFrame(outlier_info).T
print("\n🔍 Outlier Detection Results (IQR Method):")
print(outlier_df.round(2))

# Visualize outliers using box plots
fig, axes = plt.subplots(2, 4, figsize=(18, 10))
axes = axes.flatten()

for i, col in enumerate(numerical_cols):
    sns.boxplot(data=train_data, y=col, ax=axes[i], color='steelblue')
    axes[i].set_title(f'{col}', fontsize=11)
    axes[i].set_ylabel('')

plt.suptitle('Box Plots for Outlier Detection', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Decision on outlier treatment
print("\n📌 Outlier Treatment Decision:")
print("For manufacturing data, outliers often represent extreme but valid operating conditions.")
print("Since failures often occur at extremes, we should NOT remove outliers.")
print("👉 Will use robust scaling or tree-based models that handle outliers naturally.")
````

````python
# ==================================================
# 6.9 HANDLING IMBALANCED DATASET
# ==================================================

print("="*50)
print("📊 6.9: HANDLING CLASS IMBALANCE")
print("="*50)

# Check imbalance ratio
n_non_failure = (y_train == 0).sum()
n_failure = (y_train == 1).sum()
imbalance_ratio = n_non_failure / n_failure

print(f"📌 Training Set Class Distribution:")
print(f"   Non-Failure (0): {n_non_failure}")
print(f"   Failure (1): {n_failure}")
print(f"   Imbalance Ratio: {imbalance_ratio:.1f}:1")
print(f"   Failure Rate: {n_failure/(n_non_failure + n_failure)*100:.2f}%")

from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTETomek

# Method 1: SMOTE (Synthetic Minority Over-sampling)
print("\n🔹 Applying SMOTE...")
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print(f"\n✅ After SMOTE:")
print(f"   Non-Failure: {(y_train_smote == 0).sum()}")
print(f"   Failure: {(y_train_smote == 1).sum()}")
print(f"   Total samples: {len(y_train_smote)}")

# Method 2: SMOTE + Tomek Links (for comparison)
print("\n🔹 Applying SMOTETomek...")
smote_tomek = SMOTETomek(random_state=42)
X_train_st, y_train_st = smote_tomek.fit_resample(X_train, y_train)

print(f"\n✅ After SMOTETomek:")
print(f"   Non-Failure: {(y_train_st == 0).sum()}")
print(f"   Failure: {(y_train_st == 1).sum()}")
print(f"   Total samples: {len(y_train_st)}")

# Visualize before and after
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Before balancing
axes[0].bar(['Non-Failure', 'Failure'], [n_non_failure, n_failure],
            color=['#2ecc71', '#e74c3c'])
axes[0].set_title(f'Original (Imbalanced)\nRatio {imbalance_ratio:.1f}:1', fontsize=12)
axes[0].set_ylabel('Count')

# After SMOTE
axes[1].bar(['Non-Failure', 'Failure'],
            [(y_train_smote == 0).sum(), (y_train_smote == 1).sum()],
            color=['#2ecc71', '#e74c3c'])
axes[1].set_title(f'After SMOTE\n(Balanced)', fontsize=12)

# After SMOTETomek
axes[2].bar(['Non-Failure', 'Failure'],
            [(y_train_st == 0).sum(), (y_train_st == 1).sum()],
            color=['#2ecc71', '#e74c3c'])
axes[2].set_title(f'After SMOTETomek\n(Balanced)', fontsize=12)

plt.suptitle('Class Imbalance Handling Techniques', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Decision on which technique to use
print("\n" + "="*50)
print("📌 IMBALANCE HANDLING DECISION")
print("="*50)
print("✅ Will use SMOTE because:")
print("  • Creates synthetic samples (not just duplicates)")
print("  • Preserves more information than undersampling")
print("  • Faster than SMOTETomek with similar performance")
print("  • Standard choice for imbalanced classification")
print("\n👉 Will train models on both original and SMOTE data")
print("   and compare performance.")

# Store balanced data for modeling
X_train_balanced = X_train_smote
y_train_balanced = y_train_smote
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ==================================================
# ML MODEL - 1: RANDOM FOREST - FAST PIPELINE TUNING
# ==================================================

print("="*50)
print("🌲 MODEL 1: RANDOM FOREST - FAST TUNING")
print("="*50)

from imblearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
import time

# Create pipeline
pipeline = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ('rf', RandomForestClassifier(random_state=42, n_jobs=-1))
])

# Define parameter distribution (wide range but random search)
param_dist = {
    'rf__n_estimators': [50, 100, 150],
    'rf__max_depth': [5, 10, 15, None],
    'rf__min_samples_split': [2, 5, 10],
    'rf__min_samples_leaf': [1, 2, 4],
    'rf__class_weight': ['balanced', None]
}

print(f"\n🔹 Total combinations possible: {3 * 4 * 3 * 3 * 2} = 216 combinations")
print(f"🔹 Using RandomizedSearchCV with 10 iterations")
print(f"🔹 2-fold CV → {10 * 2} = 20 total fits")
print("\n⏳ This will take 2-3 minutes...")

start_time = time.time()

# Use RandomizedSearchCV instead of GridSearchCV
random_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_dist,
    n_iter=10,  # Only try 10 random combinations
    cv=2,       # Only 2 folds
    scoring='recall',
    random_state=42,
    n_jobs=-1,
    verbose=1
)

# Use 30% of training data for faster training
sample_size = int(0.3 * len(X_train))
X_sample = X_train[:sample_size]
y_sample = y_train[:sample_size]

print(f"🔹 Using {sample_size} samples (30% of data)")
random_search.fit(X_sample, y_sample)

end_time = time.time()
print(f"\n✅ Tuning completed in {(end_time - start_time)/60:.2f} minutes")

# Best parameters
print("\n" + "="*50)
print("📊 BEST PARAMETERS FOUND")
print("="*50)
print(random_search.best_params_)

# Evaluate on full validation set
best_pipeline = random_search.best_estimator_
y_pred_tuned = best_pipeline.predict(X_val)
y_pred_proba_tuned = best_pipeline.predict_proba(X_val)[:, 1]

# Calculate metrics
accuracy_tuned = accuracy_score(y_val, y_pred_tuned)
precision_tuned = precision_score(y_val, y_pred_tuned)
recall_tuned = recall_score(y_val, y_pred_tuned)
f1_tuned = f1_score(y_val, y_pred_tuned)

print("\n📊 Tuned Model Performance:")
print(f"   Accuracy:  {accuracy_tuned:.4f}")
print(f"   Precision: {precision_tuned:.4f}")
print(f"   Recall:    {recall_tuned:.4f}")
print(f"   F1-Score:  {f1_tuned:.4f}")

# Compare with original
print("\n" + "="*50)
print("📊 COMPARISON WITH ORIGINAL MODEL")
print("="*50)
print(f"Original Recall: {recall_rf:.4f}")
print(f"Tuned Recall:    {recall_tuned:.4f}")

if recall_tuned > recall_rf:
    print(f"\n✅ Improvement: +{recall_tuned - recall_rf:.4f}")
else:
    print(f"\n⚠️ No improvement. Using original model.")
````

It's faster than GridSearchCV (tested 10 random combinations out of 216 possible)

I optimized for recall as it's the most critical metric for failure prediction

Used 30% data sample and 2-fold CV to balance speed and accuracy

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ==================================================
# ML MODEL - 2: XGBOOST - HYPERPARAMETER TUNING
# ==================================================

print("="*50)
print("🚀 MODEL 2: XGBOOST - HYPERPARAMETER TUNING")
print("="*50)

from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
import time

# Prepare cleaned data (as done before)
X_train_clean = X_train_balanced.copy()
X_val_clean = X_val.copy()
X_train_clean.columns = [col.replace(' ', '_').replace('[', '').replace(']', '') for col in X_train_clean.columns]
X_val_clean.columns = [col.replace(' ', '_').replace('[', '').replace(']', '') for col in X_val_clean.columns]

# Define parameter distribution for XGBoost
param_dist_xgb = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'min_child_weight': [1, 3, 5]
}

print(f"\n🔹 Total combinations possible: {3 * 3 * 3 * 2 * 2 * 3} = 324 combinations")
print(f"🔹 Using RandomizedSearchCV with 15 iterations")
print(f"🔹 3-fold CV → {15 * 3} = 45 total fits")
print("\n⏳ This will take 3-5 minutes...")

# Create base XGBoost model
xgb_base = XGBClassifier(
    scale_pos_weight=62.5,  # Handle imbalance
    random_state=42,
    n_jobs=-1,
    eval_metric='logloss',
    use_label_encoder=False
)

# Use RandomizedSearchCV
random_search_xgb = RandomizedSearchCV(
    estimator=xgb_base,
    param_distributions=param_dist_xgb,
    n_iter=15,  # Try 15 random combinations
    cv=3,
    scoring='recall',  # Optimize for recall
    random_state=42,
    n_jobs=-1,
    verbose=1
)

# Use 40% of data for faster tuning
sample_size = int(0.4 * len(X_train_clean))
X_sample = X_train_clean[:sample_size]
y_sample = y_train_balanced[:sample_size]

print(f"🔹 Using {sample_size} samples (40% of data)")
start_time = time.time()

random_search_xgb.fit(X_sample, y_sample)

end_time = time.time()
print(f"\n✅ Tuning completed in {(end_time - start_time)/60:.2f} minutes")

# Best parameters
print("\n" + "="*50)
print("📊 BEST PARAMETERS FOUND")
print("="*50)
print(random_search_xgb.best_params_)

print(f"\n📊 Best Cross-Validation Recall: {random_search_xgb.best_score_:.4f}")

# Evaluate best model on full validation set
best_xgb = random_search_xgb.best_estimator_
y_pred_tuned_xgb = best_xgb.predict(X_val_clean)
y_pred_proba_tuned_xgb = best_xgb.predict_proba(X_val_clean)[:, 1]

# Calculate metrics
accuracy_tuned_xgb = accuracy_score(y_val, y_pred_tuned_xgb)
precision_tuned_xgb = precision_score(y_val, y_pred_tuned_xgb)
recall_tuned_xgb = recall_score(y_val, y_pred_tuned_xgb)
f1_tuned_xgb = f1_score(y_val, y_pred_tuned_xgb)

print("\n📊 Tuned XGBoost Performance:")
print(f"   Accuracy:  {accuracy_tuned_xgb:.4f}")
print(f"   Precision: {precision_tuned_xgb:.4f}")
print(f"   Recall:    {recall_tuned_xgb:.4f}")
print(f"   F1-Score:  {f1_tuned_xgb:.4f}")

# Compare with original XGBoost
print("\n" + "="*50)
print("📊 COMPARISON: ORIGINAL vs TUNED XGBoost")
print("="*50)
print(f"Original XGBoost Recall: {recall_xgb:.4f}")
print(f"Tuned XGBoost Recall:    {recall_tuned_xgb:.4f}")

if recall_tuned_xgb > recall_xgb:
    print(f"\n✅ Improvement: +{recall_tuned_xgb - recall_xgb:.4f}")
else:
    print(f"\n⚠️ No improvement. Using original XGBoost model.")
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ==================================================
# ML MODEL - 3: LOGISTIC REGRESSION - HYPERPARAMETER TUNING
# ==================================================

print("="*50)
print("📈 MODEL 3: LOGISTIC REGRESSION - HYPERPARAMETER TUNING")
print("="*50)

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import time

# Define parameter grid for Logistic Regression
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],  # Regularization strength
    'penalty': ['l1', 'l2'],    # Regularization type
    'solver': ['liblinear', 'saga']  # Algorithms that support both penalties
}

print(f"\n🔹 Total combinations: {4 * 2 * 2} = 16 combinations")
print(f"🔹 Using GridSearchCV with 3-fold CV → {16 * 3} = 48 total fits")
print("\n⏳ This will take 1-2 minutes...")

# Create base Logistic Regression model
lr_base = LogisticRegression(
    class_weight='balanced',
    random_state=42,
    max_iter=1000,
    n_jobs=-1
)

# Use GridSearchCV
grid_search_lr = GridSearchCV(
    estimator=lr_base,
    param_grid=param_grid_lr,
    cv=3,
    scoring='recall',  # Optimize for recall
    n_jobs=-1,
    verbose=1
)

# Use 60% of data for faster tuning
sample_size = int(0.6 * len(X_train_balanced))
X_sample = X_train_balanced[:sample_size]
y_sample = y_train_balanced[:sample_size]

print(f"🔹 Using {sample_size} samples (60% of data)")
start_time = time.time()

grid_search_lr.fit(X_sample, y_sample)

end_time = time.time()
print(f"\n✅ Tuning completed in {(end_time - start_time)/60:.2f} minutes")

# Best parameters
print("\n" + "="*50)
print("📊 BEST PARAMETERS FOUND")
print("="*50)
print(grid_search_lr.best_params_)

print(f"\n📊 Best Cross-Validation Recall: {grid_search_lr.best_score_:.4f}")

# Evaluate best model on full validation set
best_lr = grid_search_lr.best_estimator_
y_pred_tuned_lr = best_lr.predict(X_val)
y_pred_proba_tuned_lr = best_lr.predict_proba(X_val)[:, 1]

# Calculate metrics
accuracy_tuned_lr = accuracy_score(y_val, y_pred_tuned_lr)
precision_tuned_lr = precision_score(y_val, y_pred_tuned_lr)
recall_tuned_lr = recall_score(y_val, y_pred_tuned_lr)
f1_tuned_lr = f1_score(y_val, y_pred_tuned_lr)

print("\n📊 Tuned Logistic Regression Performance:")
print(f"   Accuracy:  {accuracy_tuned_lr:.4f}")
print(f"   Precision: {precision_tuned_lr:.4f}")
print(f"   Recall:    {recall_tuned_lr:.4f}")
print(f"   F1-Score:  {f1_tuned_lr:.4f}")

# Compare with original Logistic Regression
print("\n" + "="*50)
print("📊 COMPARISON: ORIGINAL vs TUNED LOGISTIC REGRESSION")
print("="*50)
print(f"Original Logistic Regression Recall: {recall_lr:.4f}")
print(f"Tuned Logistic Regression Recall:    {recall_tuned_lr:.4f}")

if recall_tuned_lr > recall_lr:
    print(f"\n✅ Improvement: +{recall_tuned_lr - recall_lr:.4f}")
else:
    print(f"\n⚠️ No improvement. Using original Logistic Regression model.")

# Final comparison of all tuned models
print("\n" + "="*50)
print("📊 FINAL TUNED MODELS COMPARISON")
print("="*50)

final_tuned_comparison = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'Random Forest (Tuned)': [accuracy_tuned, precision_tuned, recall_tuned, f1_tuned],
    'XGBoost (Original)': [accuracy_xgb, precision_xgb, recall_xgb, f1_xgb],
    'Logistic Regression (Tuned)': [accuracy_tuned_lr, precision_tuned_lr, recall_tuned_lr, f1_tuned_lr]
})

print(final_tuned_comparison.round(4))

# Final recommendation
print("\n" + "="*50)
print("📌 FINAL PRODUCTION MODEL RECOMMENDATION")
print("="*50)

models_recall = [
    ('Random Forest', recall_tuned),
    ('XGBoost', recall_xgb),
    ('Logistic Regression', recall_tuned_lr)
]
best_model = max(models_recall, key=lambda x: x[1])

print(f"✅ BEST MODEL: {best_model[0]} with {best_model[1]:.2%} recall")
print("\n👉 Recommended for TATA Steel production deployment:")
print(f"   • Model: {best_model[0]}")
print(f"   • Recall: {best_model[1]:.2%} (catches {int(best_model[1] * 430)} out of 430 failures)")
print(f"   • Missed Failures: {430 - int(best_model[1] * 430)}")
````
